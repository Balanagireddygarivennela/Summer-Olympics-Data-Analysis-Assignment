2. Which sport is having more number of GoldMedals

```python
Gold_Medals = df[df['Medal']=='Gold']
```

```python
Gold_Medals.groupby('Sport').Medal.count().sort_values(ascending = False).head()
```

```python
Gold_Medals.groupby('Sport').Medal.count().sort_values(ascending = False).head().plot.bar(figsize=(6,5))
```

```python
Gold_Medals.Sport.value_counts().sort_values(ascending = False).head().plot.bar(figsize=(6,5))
```

3.Which Sport is having most number of medals so far?(Top 5)

```python
df.groupby('Sport').Medal.count().sort_values(ascending = False).head().plot.bar(figsize = (6,5))
```
